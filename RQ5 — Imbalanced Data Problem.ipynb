# RQ5 - Model Ranking

```
import pandas as pd
import matplotlib.pyplot as plt

data = {"Model":["XGB","RF","DT","LR"],"Rank":[1,2,3,4]}
df = pd.DataFrame(data)
df.to_csv("RQ5_table.csv", index=False)

plt.plot(df["Model"], df["Rank"], marker='o')
plt.title("Model Ranking")
plt.savefig("RQ5_figure.pdf")
plt.show()
```
