# RQ6 - Robustness

```
import pandas as pd
import matplotlib.pyplot as plt

data = {"Scenario":["Split","CV","Noise","Missing"],"R2":[0.86,0.85,0.81,0.78]}
df = pd.DataFrame(data)
df.to_csv("RQ6_table.csv", index=False)

plt.plot(df["Scenario"], df["R2"], marker='o')
plt.title("Robustness Analysis")
plt.savefig("RQ6_figure.pdf")
plt.show()
```
